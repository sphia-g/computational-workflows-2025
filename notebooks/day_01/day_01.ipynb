1. What is nf-core?

nf-core hosts Nextflow pipelines for biosciences. It focuses on ensuring reproducibility and portability of pipelines across different hardware, operating systems, and software versions.

2. How many pipelines are there currently in nf-core?

139 pipelines

3. Are there any non-bioinformatic pipelines in nf-core?

Yes, there are e.g. general data processing or infrastructure-related workflows

4. Let's go back a couple of steps. What is a pipeline and what do we use it for?

A pipeline is a set of data processing elements connected in series, where the output of one element is the input of the next one. We would use them to save time, reduce errors, and make analyses reproducible and shareable.

5. Why do you think nf-core adheres to strict guidelines?

To ensure reproducibility, maintainability and incrrease trust in the developed pipelines

6. What are the main features of nf-core pipelines?

Sharable pipelines, comunity-driven, standardized and portable, modular
